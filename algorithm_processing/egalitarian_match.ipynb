```python
from stable_matching import generate_responses, generate_prefrence_list, ppdictionary, flatten_preference_dictionary
```

```python
print("\nStable Matching Hetero Demo:\n")

male = generate_responses(NUMBER_OF_PEOPLE=100, NUMBER_OF_CATEGORIES=5, NUMBER_OF_QUESTIONS=15)

female = generate_responses(NUMBER_OF_PEOPLE=100, NUMBER_OF_CATEGORIES=5, NUMBER_OF_QUESTIONS=15)

male_preference, female_preference = generate_prefrence_list(male, female)

md, fd = flatten_preference_dictionary(male_preference, female_preference)

print("Male Preference Table")

ppdictionary(md)

print("Female Preference Table")

ppdictionary(fd)
```

```python
from matching_algorithms import egalitarian_stable_matching
# Egalitarian Matching For Hetero Populations
male_pref = md
fema_pref = fd
matching = egalitarian_stable_matching(male_pref, fema_pref)
print("Egalitarian Stable Matching:")
for man, woman in matching.items():
    print(f"{man} - {woman}")
```
